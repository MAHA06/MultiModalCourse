# Parameter estimation: curve fitting

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def func(x): return np.sin(2*np.pi*x)

# Generate sinusoidal data with noise
size = 25
rng = np.random.RandomState(1234)
x = rng.uniform(0., 1., size)
y = func(x) + rng.normal(scale=0.1, size=size)
xt = np.linspace(0., 1., 100)

# Generate sinusoidal data with noise and outliers
size = 25
bias = 2.5
rng = np.random.RandomState(1234)
x = rng.uniform(0., 1., size)
y2 = func(x) + rng.normal(scale=0.1, size=size)
# Add outlier bias
y2[15:20] = y2[15:20] + bias

f = plt.figure(figsize=(10,4))
f.add_subplot(1,2,1)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')

f.add_subplot(1,2,2)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y2, '*')
```

## Linear regression

```python
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

# Create linear regression object
lreg = linear_model.LinearRegression()
# Fit the model to data
n_order = 3
X = PolynomialFeatures(n_order).fit_transform(np.array([x]).T)
Xt = PolynomialFeatures(n_order).fit_transform(np.array([xt]).T)

lreg.fit(X, y)
print('parameters:', lreg.coef_)
ypred = lreg.predict(Xt)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')
plt.plot(xt, ypred, 'r')
```

## Bayesian linear regression

```python
breg = linear_model.BayesianRidge(tol=1e-6, fit_intercept=False, compute_score=True)

breg.fit(X, y)
print('parameters:', breg.coef_)

f = plt.figure(figsize=(10,3))

ymean, ystd = breg.predict(Xt, return_std=True)

f.add_subplot(1,2,1)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')
plt.plot(xt, ymean, 'r')
plt.fill_between(xt, ymean-ystd, ymean+ystd,
                    color="pink", alpha=0.5, label="predict std")

breg.set_params(alpha_init=1., lambda_init= 1e-3)

breg.fit(X, y)
print('parameters:', breg.coef_)

f.add_subplot(1,2,2)
ymean, ystd = breg.predict(Xt, return_std=True)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')
plt.plot(xt, ymean, 'r')
plt.fill_between(xt, ymean-2*ystd, ymean+2*ystd,
                    color="pink", alpha=0.5, label="predict std")
```

# Robust parameter estimation: curve fitting

```python
# Generate sinusoidal data with noise and outliers
size = 25
bias = 2.5
rng = np.random.RandomState(1234)
x = rng.uniform(0., 1., size)
y = func(x) + rng.normal(scale=0.1, size=size)
# Add outlier bias
y[15:20] = y[15:20] + bias
xt = np.linspace(0., 1., 100)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')
```

## Bayesian linear regression

```python
lreg.fit(X, y)
print('parameters:', lreg.coef_)
ypred = lreg.predict(Xt)

f = plt.figure(figsize=(15,4))

f.add_subplot(1,3,1)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')
plt.plot(xt, ypred, 'r')

breg = linear_model.BayesianRidge(tol=1e-6, fit_intercept=False, compute_score=True)

breg.fit(X, y)
print('parameters:', breg.coef_)

ymean, ystd = breg.predict(Xt, return_std=True)

f.add_subplot(1,3,2)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')
plt.plot(xt, ymean, 'r')
plt.fill_between(xt, ymean-ystd, ymean+ystd,
                    color="pink", alpha=0.5, label="predict std")

breg.set_params(alpha_init=1., lambda_init= 1e-3)

breg.fit(X, y)
print('parameters:', breg.coef_)

f.add_subplot(1,3,3)
ymean, ystd = breg.predict(Xt, return_std=True)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')
plt.plot(xt, ymean, 'r')
plt.fill_between(xt, ymean-2*ystd, ymean+2*ystd,
                    color="pink", alpha=0.5, label="predict std")
```

## RANSAC

```python
# Create linear regression object
ransac = linear_model.RANSACRegressor()
ransac.fit(X, y)

ypred = ransac.predict(Xt)
plt.plot(xt, func(xt), 'b')
plt.plot(x, y, '*')
plt.plot(xt, ypred, 'r')
```

## Gaussian Mixture Model

```python
from sklearn.mixture import GaussianMixture
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv("bananas.csv",skiprows=0)
df = pd.read_csv("gmm_data.csv")
plt.scatter(df.iloc[:, 0],df.iloc[:, 1])
```

```python
X = np.array(df.iloc[:,0:2])

labels = KMeans(4).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

```python
gmm = GaussianMixture(n_components=4,covariance_type='full',random_state = 100)
labels2 = gmm.fit(X).predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels2)
plt.show()
```

```python
probs = gmm.predict_proba(X)
print(probs[:5].round(2))
```
